# Encodeing the categorical variable to numerical transformations

```python
lable_encoder=LabelEncoder()
data_frame['department']=lable_encoder.fit_transform(data_frame['department'])
data_frame['region']=lable_encoder.fit_transform(data_frame['region'])
data_frame['education']=lable_encoder.fit_transform(data_frame['education'])
data_frame['gender']=lable_encoder.fit_transform(data_frame['gender'])
data_frame['recruitment_channel']=lable_encoder.fit_transform(data_frame['recruitment_channel'])
```
